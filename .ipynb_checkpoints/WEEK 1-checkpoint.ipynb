```python
# Count the number of occurrences per Celeb_ID
id_counts = df['celeb id'].value_counts().sort_index()

# Convert to DataFrame if you want tabular format
counts_df = id_counts.reset_index()
counts_df.columns = ['celeb id', 'Count']

counts_df
```

```python
# Step 1: Get the top 50 Celeb_IDs
top_50_ids = counts_df.sort_values(by="Count", ascending=False).head(50)["celeb id"]

# Step 2: Filter original dataframe for only those IDs
filtered_df = df[df["celeb id"].isin(top_50_ids)]

# Step 3: Group by Celeb_ID and collect image filenames into a list
celeb_dict = filtered_df.groupby("celeb id")["image id"].apply(list).to_dict()

# celeb_dict will now be a dictionary with structure:
# { Celeb_ID: [list of images], ... }
celeb_dict
```

```python
# ====================== 2) Copy images for TOP-50 ======================
import os, shutil
from pathlib import Path

# Original images folder (try both single and double nesting)
base = Path("/Users/nishanthmano3112/Desktop/FALL 2025/DL & NN (IE 7615)/Discriminative Project /CelebA Proj/CelebAdata")
img_dir_candidates = [
    base / "img_align_celeba" / "img_align_celeba",
    base / "img_align_celeba",
]
image_folder = None
for cand in img_dir_candidates:
    if cand.exists():
        image_folder = cand
        break
assert image_folder is not None, "Could not find img_align_celeba folder."

# Output folder JUST for the top-50 classes (staging before split)
output_folder = base / "Dataset 50"
output_folder.mkdir(parents=True, exist_ok=True)

# Copy
for celeb_id, img_list in celeb_dict.items():
    celeb_folder = output_folder / str(celeb_id)
    celeb_folder.mkdir(parents=True, exist_ok=True)
    for img_file in img_list:
        src_path = image_folder / img_file
        if src_path.exists():
            shutil.copy(src_path, celeb_folder / img_file)
        else:
            # Fallback search in case of odd extraction
            hits = list(image_folder.rglob(img_file))
            if hits:
                shutil.copy(hits[0], celeb_folder / img_file)
            else:
                print(f"[WARN] Missing: {src_path}")
print("Copied images for top-50 into:", output_folder)
```

```python
# ====================== 3) 70/15/15 split -> .../data/train|val|test ======================
import os, shutil, random
from pathlib import Path

# Top-50 staged folder:
DATA = Path("/Users/nishanthmano3112/Desktop/FALL 2025/DL & NN (IE 7615)/Discriminative Project /CelebA Proj/CelebAdata/Dataset 50")

# Final split root that your training code expects:
DATA_ROOT = Path("/Users/nishanthmano3112/Desktop/FALL 2025/DL & NN (IE 7615)/Discriminative Project /CelebA Proj/CelebAdata/data")
TRAIN = DATA_ROOT / "train"
VAL   = DATA_ROOT / "val"
TEST  = DATA_ROOT / "test"

SPLIT = (0.70, 0.15, 0.15)  # train/val/test
SEED  = 42
RESET_OUTPUT = True
IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".webp", ".tif", ".tiff"}

random.seed(SEED)
if RESET_OUTPUT:
    for p in [TRAIN, VAL, TEST]:
        if p.exists():
            shutil.rmtree(p)
for p in [TRAIN, VAL, TEST]:
    p.mkdir(parents=True, exist_ok=True)

def split_counts(n, ratios=(0.70, 0.15, 0.15)):
    if n <= 0: return 0, 0, 0
    if n == 1: return 1, 0, 0
    if n == 2: return 1, 0, 1
    t = round(n * ratios[0])
    v = round(n * ratios[1])
    te = n - t - v
    if v == 0:
        v = 1; t = max(1, t-1); te = n - t - v
    if te == 0:
        te = 1
        if t > 1: t -= 1
        else:     v = max(1, v-1)
        te = n - t - v
    return t, v, te

total_train = total_val = total_test = 0
classes = 0

for celeb_id in os.listdir(DATA):
    src_dir = DATA / celeb_id
    if not src_dir.is_dir():
        continue

    imgs = [f for f in os.listdir(src_dir)
            if not f.startswith(".") and Path(f).suffix.lower() in IMG_EXTS]
    if not imgs:
        continue

    random.shuffle(imgs)
    n = len(imgs)
    n_train, n_val, n_test = split_counts(n, SPLIT)

    train_imgs = imgs[:n_train]
    val_imgs   = imgs[n_train:n_train+n_val]
    test_imgs  = imgs[n_train+n_val:]

    (TRAIN/celeb_id).mkdir(parents=True, exist_ok=True)
    (VAL/celeb_id).mkdir(parents=True, exist_ok=True)
    (TEST/celeb_id).mkdir(parents=True, exist_ok=True)

    for f in train_imgs: shutil.copy(src_dir/f, TRAIN/celeb_id/f)
    for f in val_imgs:   shutil.copy(src_dir/f, VAL/celeb_id/f)
    for f in test_imgs:  shutil.copy(src_dir/f, TEST/celeb_id/f)

    total_train += len(train_imgs)
    total_val   += len(val_imgs)
    total_test  += len(test_imgs)
    classes     += 1

print(f"Done. Classes processed: {classes} (expect ~50)")
print(f"Train: {total_train} | Val: {total_val} | Test: {total_test}")
print("Output roots:\n ", TRAIN, "\n ", VAL, "\n ", TEST)
```

```python
# ===== ResNet50 (grayscale) with Train/Val/Test, best-on-val checkpoint =====
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torchvision.models import ResNet50_Weights
from torch.utils.data import DataLoader

# ----------------------------
# 0) Paths & device
# ----------------------------
data_dir = '/Users/nishanthmano3112/Desktop/FALL 2025/DL & NN (IE 7615)/Discriminative Project /CelebA Proj/CelebAdata/data'

if torch.cuda.is_available():
    device = torch.device("cuda")
elif hasattr(torch.backends, "mps") and torch.backends.mps.is_available():
    device = torch.device("mps")
else:
    device = torch.device("cpu")
print("Using device:", device)

# ----------------------------
# 1) Transforms & Datasets (Grayscale)
# ----------------------------
train_tf = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.Grayscale(num_output_channels=1),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

eval_tf = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.Grayscale(num_output_channels=1),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

train_dataset = datasets.ImageFolder(root=os.path.join(data_dir, "train"), transform=train_tf)
val_dataset   = datasets.ImageFolder(root=os.path.join(data_dir, "val"),   transform=eval_tf)
test_dataset  = datasets.ImageFolder(root=os.path.join(data_dir, "test"),  transform=eval_tf)

assert train_dataset.classes == val_dataset.classes == test_dataset.classes, \
    "Train/Val/Test must have identical class folders."

num_classes = len(train_dataset.classes)
print(f"Number of classes: {num_classes}")

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True,  num_workers=0)
val_loader   = DataLoader(val_dataset,   batch_size=32, shuffle=False, num_workers=0)
test_loader  = DataLoader(test_dataset,  batch_size=32, shuffle=False, num_workers=0)

# ----------------------------
# 2) Model: ResNet50 (ImageNet-pretrained → grayscale conv)
# ----------------------------
def _avg_first_conv_weight_to_gray(w_rgb: torch.Tensor):
    return w_rgb.mean(dim=1, keepdim=True)

def get_resnet50(num_classes: int):
    model = models.resnet50(weights=ResNet50_Weights.IMAGENET1K_V1)
    # adapt first conv to 1 channel with averaged RGB weights
    w = model.conv1.weight.data.clone()
    model.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
    with torch.no_grad():
        model.conv1.weight[:] = _avg_first_conv_weight_to_gray(w)
    # replace classifier
    model.fc = nn.Linear(model.fc.in_features, num_classes)
    return model

model = get_resnet50(num_classes).to(device)

# ----------------------------
# 3) Accuracy function
# ----------------------------
def accuracy(loader, model, device):
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            logits = model(x)
            pred = logits.argmax(1)
            correct += (pred == y).sum().item()
            total += y.size(0)
    return 100.0 * correct / max(1, total)

# ----------------------------
# 4) Training (Test only at end)
# ----------------------------
def train_and_evaluate(model, epochs=30, lr=1e-3, weight_decay=1e-3, ckpt_dir="./checkpoints"):
    os.makedirs(ckpt_dir, exist_ok=True)
    best_ckpt = os.path.join(ckpt_dir, "ResNet50_best.pth")
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)

    best_val = -1.0
    history = {"loss": [], "train_acc": [], "val_acc": []}

    for ep in range(1, epochs+1):
        model.train()
        running_loss, correct, total = 0.0, 0, 0

        for x, y in train_loader:
            x, y = x.to(device), y.to(device)
            optimizer.zero_grad(set_to_none=True)
            logits = model(x)
            loss = criterion(logits, y)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            correct += (logits.argmax(1) == y).sum().item()
            total += y.size(0)

        train_acc = 100.0 * correct / max(1, total)
        avg_loss  = running_loss / max(1, len(train_loader))
        val_acc   = accuracy(val_loader, model, device)

        # Save best checkpoint based on validation accuracy
        if val_acc > best_val:
            best_val = val_acc
            torch.save({"model_state": model.state_dict(),
                        "num_classes": num_classes}, best_ckpt)

        history["loss"].append(avg_loss)
        history["train_acc"].append(train_acc)
        history["val_acc"].append(val_acc)

        print(f"Epoch {ep:02d}/{epochs} | "
              f"Loss {avg_loss:.4f} | Train {train_acc:.2f}% | "
              f"Val {val_acc:.2f}% | Best Val {best_val:.2f}%")

    # ✅ Only run test once at the very end
    if os.path.exists(best_ckpt):
        ckpt = torch.load(best_ckpt, map_location=device)
        model.load_state_dict(ckpt["model_state"])
        final_test = accuracy(test_loader, model, device)
        print(f"\n✅ ResNet50 | Final TEST (best-on-val checkpoint): {final_test:.2f}%")

    return history

# ----------------------------
# 5) Run
# ----------------------------
history = train_and_evaluate(model, epochs=30, lr=1e-3, weight_decay=1e-3, ckpt_dir="./checkpoints")

# ----------------------------
# 6) Plot metrics
# ----------------------------
def plot_history(history, title="ResNet50 Training History"):
    epochs = range(1, len(history["loss"]) + 1)

    plt.figure(figsize=(12,5))

    # Plot Loss
    plt.subplot(1,2,1)
    plt.plot(epochs, history["loss"], label="Train Loss", color="red")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training Loss")
    plt.grid(True)
    plt.legend()

    # Plot Accuracy
    plt.subplot(1,2,2)
    plt.plot(epochs, history["train_acc"], label="Train Acc", marker="o")
    plt.plot(epochs, history["val_acc"],   label="Val Acc",   marker="s")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy (%)")
    plt.title("Training & Validation Accuracy")
    plt.grid(True)
    plt.legend()

    plt.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()

# After training:
history = train_and_evaluate(model, epochs=30, lr=1e-3, weight_decay=1e-3, ckpt_dir="./checkpoints")
plot_history(history)
```

```python
from torchvision import models

def _to_gray_first_conv(module):
    # average RGB weights to 1-channel conv if needed
    if hasattr(module, "weight") and module.weight.ndim==4 and module.weight.shape[1]==3:
        with torch.no_grad():
            w = module.weight.data
            module.in_channels = 1
            new = torch.nn.Conv2d(1, w.shape[0], kernel_size=module.kernel_size,
                                  stride=module.stride, padding=module.padding, bias=False)
            new.weight[:] = w.mean(dim=1, keepdim=True)
        return new
    return module

def get_densenet121(num_classes:int):
    m = models.densenet121(weights=models.DenseNet121_Weights.IMAGENET1K_V1)
    # swap first conv to 1-ch
    m.features.conv0 = _to_gray_first_conv(m.features.conv0)
    # classifier
    m.classifier = nn.Linear(m.classifier.in_features, num_classes)
    return m

def train_ce(model, epochs=30, lr=1e-3, wd=1e-3, ckpt='./ckpts_densenet'):
    os.makedirs(ckpt, exist_ok=True)
    best_path = os.path.join(ckpt, 'best.pth')
    crit = nn.CrossEntropyLoss()
    opt  = optim.Adam(model.parameters(), lr=lr, weight_decay=wd)
    best_val = -1.0
    for ep in range(1, epochs+1):
        model.train()
        run_loss=cor=tot=0
        for xb,yb in train_loader:
            xb,yb = xb.to(device), yb.to(device)
            opt.zero_grad(set_to_none=True)
            logits = model(xb)
            loss = crit(logits, yb)
            loss.backward()
            opt.step()
            run_loss += loss.item()
            cor += (logits.argmax(1)==yb).sum().item()
            tot += yb.size(0)
        tr_acc = 100.0*cor/max(1,tot)
        va_acc = accuracy(val_loader, model, device)
        if va_acc > best_val:
            best_val = va_acc
            torch.save({"model":model.state_dict(),"num_classes":num_classes}, best_path)
        print(f"[DenseNet121] Epoch {ep:02d}/30 | Loss {run_loss/len(train_loader):.4f} | "
              f"Train {tr_acc:.2f}% | Val {va_acc:.2f}% | Best {best_val:.2f}%")
    if os.path.exists(best_path):
        ckpt = torch.load(best_path, map_location=device)
        model.load_state_dict(ckpt["model"])
    te = accuracy(test_loader, model, device)
    print(f"✅ DenseNet121 TEST: {te:.2f}%")
    return model

densenet = get_densenet121(num_classes).to(device)
_ = train_ce(densenet, epochs=30, lr=1e-3, wd=1e-3, ckpt='./ckpts_densenet')

# ------------------------
# Plotting
# ------------------------
def plot_history(history, title="DenseNet121 Training History"):
    epochs = range(1, len(history["loss"]) + 1)
    plt.figure(figsize=(12,5))

    # Loss
    plt.subplot(1,2,1)
    plt.plot(epochs, history["loss"], label="Train Loss", color="red")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training Loss")
    plt.grid(True)
    plt.legend()

    # Accuracy
    plt.subplot(1,2,2)
    plt.plot(epochs, history["train_acc"], label="Train Acc", marker="o")
    plt.plot(epochs, history["val_acc"],   label="Val Acc",   marker="s")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy (%)")
    plt.title("Training vs Validation Accuracy")
    plt.grid(True)
    plt.legend()

    plt.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()


# ------------------------
# Run training + plot
# ------------------------
densenet = get_densenet121(num_classes).to(device)
densenet, history = train_ce(densenet, epochs=30, lr=1e-3, wd=1e-3, ckpt='./ckpts_densenet')
plot_history(history, title="DenseNet121 on CelebA (Grayscale)")
```

```python
# 2) Model: ResNet50 - AFFACT adapted to grayscale
# ----------------------------
class ResNet50_AFFACT(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        m = models.resnet50(weights=ResNet50_Weights.IMAGENET1K_V1)
        # adapt conv1 to 1-channel
        w = m.conv1.weight.data.clone()
        m.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
        with torch.no_grad():
            m.conv1.weight[:] = w.mean(dim=1, keepdim=True)
        # classifier head
        m.fc = nn.Linear(m.fc.in_features, num_classes)
        self.m = m
    def forward(self, x):
        return self.m(x)

# ----------------------------
# 3) Training + History
# ----------------------------
def train_affact(model, epochs=30, lr=1e-3, wd=1e-3, ckpt='./ckpts_affact'):
    os.makedirs(ckpt, exist_ok=True)
    best_path = os.path.join(ckpt, 'best.pth')
    ce = nn.CrossEntropyLoss()
    opt = optim.Adam(model.parameters(), lr=lr, weight_decay=wd)
    best_val = -1.0

    history = {"loss": [], "train_acc": [], "val_acc": []}

    for ep in range(1, epochs+1):
        model.train()
        run_loss = cor = tot = 0
        for xb, yb in affact_train_loader:
            xb, yb = xb.to(device), yb.to(device)
            opt.zero_grad(set_to_none=True)
            logits = model(xb)
            loss = ce(logits, yb)
            loss.backward()
            opt.step()

            run_loss += loss.item()
            cor += (logits.argmax(1) == yb).sum().item()
            tot += yb.size(0)

        tr_acc = 100.0 * cor / max(1, tot)
        avg_loss = run_loss / len(affact_train_loader)
        va_acc = accuracy(val_loader, model, device)

        if va_acc > best_val:
            best_val = va_acc
            torch.save(model.state_dict(), best_path)

        history["loss"].append(avg_loss)
        history["train_acc"].append(tr_acc)
        history["val_acc"].append(va_acc)

        print(f"[AFFACT] Epoch {ep:02d}/{epochs} | Loss {avg_loss:.4f} | "
              f"Train {tr_acc:.2f}% | Val {va_acc:.2f}% | Best {best_val:.2f}%")

    if os.path.exists(best_path):
        model.load_state_dict(torch.load(best_path, map_location=device))
    te = accuracy(test_loader, model, device)
    print(f"✅ AFFACT-style TEST: {te:.2f}%")

    return model, history

# ----------------------------
# 4) Plotting
# ----------------------------
def plot_history(history, title="AFFACT Training History"):
    epochs = range(1, len(history["loss"]) + 1)
    plt.figure(figsize=(12,5))

    # Loss
    plt.subplot(1,2,1)
    plt.plot(epochs, history["loss"], label="Train Loss", color="red")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training Loss")
    plt.grid(True)
    plt.legend()

    # Accuracy
    plt.subplot(1,2,2)
    plt.plot(epochs, history["train_acc"], label="Train Acc", marker="o")
    plt.plot(epochs, history["val_acc"],   label="Val Acc",   marker="s")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy (%)")
    plt.title("Training vs Validation Accuracy")
    plt.grid(True)
    plt.legend()

    plt.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()

# ----------------------------
# 5) Run
# ----------------------------
affact = ResNet50_AFFACT(num_classes=num_classes).to(device)
affact, affact_hist = train_affact(affact, epochs=30, lr=1e-3, wd=1e-3, ckpt='./ckpts_affact')
plot_history(affact_hist, title="AFFACT (ResNet50, grayscale, strong augs)")
```
